```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


%matplotlib inline
```

```python
df = pd.read_csv("../dataset/ai4i2020_sample.csv")

df.head()
```

```python
print("Number of Rows :", df.shape[0])
print("Number of Columns :", df.shape[1])
```

```python
df.info()
```

```python
print(df.columns)
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
duplicates = df.duplicated().sum()

print("Duplicate Rows :", duplicates)
```

```python
df["Machine failure"].value_counts()
```

```python
failure = df["Machine failure"].value_counts()

plt.figure(figsize=(6,4))

plt.bar(["Healthy","Failure"],failure.values)

plt.title("Machine Failure Distribution")

plt.xlabel("Machine Status")

plt.ylabel("Number of Machines")

plt.show()
```

```python
df["Type"].value_counts()
```

```python
machine=df["Type"].value_counts()

plt.figure(figsize=(6,4))

plt.bar(machine.index,machine.values)

plt.title("Machine Type Distribution")

plt.xlabel("Machine Type")

plt.ylabel("Count")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df["Air temperature [K]"],bins=20)

plt.title("Air Temperature Distribution")

plt.xlabel("Temperature (K)")

plt.ylabel("Frequency")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df["Process temperature [K]"],bins=20)

plt.title("Process Temperature Distribution")

plt.xlabel("Temperature (K)")

plt.ylabel("Frequency")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df["Rotational speed [rpm]"],bins=20)

plt.title("Rotational Speed Distribution")

plt.xlabel("RPM")

plt.ylabel("Frequency")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df["Torque [Nm]"],bins=20)

plt.title("Torque Distribution")

plt.xlabel("Torque (Nm)")

plt.ylabel("Frequency")

plt.show()
```

```python

plt.figure(figsize=(8,5))

plt.hist(df["Tool wear [min]"],bins=20)

plt.title("Tool Wear Distribution")

plt.xlabel("Minutes")

plt.ylabel("Frequency")

plt.show()
```

```python
failure_type = df.groupby("Type")["Machine failure"].sum()

plt.figure(figsize=(6,4))

plt.bar(failure_type.index,failure_type.values)

plt.title("Failure by Machine Type")

plt.xlabel("Machine Type")

plt.ylabel("Failures")

plt.show()
```

```python
df_corr = df.copy()

df_corr["Type"] = df_corr["Type"].map({
    "L":0,
    "M":1,
    "H":2
})

corr = df_corr.drop(columns=["Product ID"]).corr()

corr
```
